```python
from datasets import load_dataset

dataset = load_dataset("maharshipandya/spotify-tracks-dataset")
```

```python
sample_size = 5000
```

```python
import pandas as pd
df = dataset["train"].to_pandas()
df = df.set_index(df.columns[0])
df.index.name = None
df = df.sample(sample_size, ignore_index=True, random_state=42).set_index("track_id")
```
